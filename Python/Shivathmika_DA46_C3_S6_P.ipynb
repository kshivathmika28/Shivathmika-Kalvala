```python
from abc import ABC, abstractmethod  # Importing ABC and abstractmethod to create an abstract base class

# Abstract base class for AppStore
class AppStore(ABC):
    # Abstract method that subclasses must implement
    @abstractmethod
    def display(self):
        pass

# Analytics class for the Analytics App
class Analytics(AppStore):
    def __init__(self, quantity):
        self.app_type = "Analytics App"  # Setting app type name
        self.quantity = quantity  # Number of apps the customer wants
        self.without_tax = 1000  # Price of one app without tax
        self.with_tax = self.without_tax * 1.18  # Price with 18% tax
        self.total_analytics = self.with_tax * self.quantity  # Total price for the quantity of apps

    def display(self):
        # Display the details for the Analytics app
        print(f"\nPurpose of the App: {self.app_type}")
        print(f"\nHow many you ordered: {self.quantity}")
        print(f"\nAmount for one {self.app_type} app is {self.with_tax} Euros")
        print(f"\nAmount for {self.quantity} is {self.total_analytics} Euros")
        print("*" * 100)
        return self.total_analytics

# Development class for the Development App
class Development(AppStore):
    def __init__(self, quantity):
        self.app_type = "Development App"  # Setting app type name
        self.quantity = quantity  # Number of apps the customer wants
        self.without_tax = 500  # Price of one app without tax
        self.with_tax = self.without_tax * 1.18  # Price with 18% tax
        self.total_development = self.with_tax * self.quantity  # Total price for the quantity of apps

    def display(self):
        # Display the details for the Development app
        print(f"\nPurpose of the App: {self.app_type}")
        print(f"\nHow many you want: {self.quantity}")
        print(f"\nAmount for one {self.app_type} app is {self.with_tax} Euros")
        print(f"\nAmount for {self.quantity} is {self.total_development} Euros")
        print("*" * 100)
        return self.total_development

# Checkout class responsible for calculating the total amount
class Checkout:
    def calculate_amount(self):
        print("*" * 100)
        # Ask the customer for how many Analytics apps they want to buy
        an_quantity = int(input("How many Analytics apps do you want to buy? "))
        an_ob = Analytics(an_quantity)  # Create an object of Analytics class
        final_analytics = an_ob.display()  # call display method and receive total amount for analytics apps

        # Ask the customer for how many Development apps they want to buy
        de_quantity = int(input("How many Development apps do you want to buy? "))
        de_ob = Development(de_quantity)  # Create an object of Development class
        final_development = de_ob.display()  # call display method and store total amount for Development apps
        
        # Calculate the final total amount
        final_amount = final_analytics + final_development
        return final_amount

# Main function for interacting with the customer
def main():
    # Ask the customer for their name
    customer_name = input("What is your name? ")
    
    print(f"\nHello {customer_name}, Welcome to The App Store-y")  # Greet the customer
    print("\nTypes of Apps available in the Store:")
    print("\nAnalytics App (price 1000 Euro per app + 18% tax)")
    print("\nDevelopment App (price 500 Euro per app + 18% tax)")
    
    # Create an object of Checkout class and store the total amount (final_analytics + final_development)
    ob_checkout = Checkout()
    to_pay = ob_checkout.calculate_amount()

    # Display the final amount to pay
    print(f"\nFinal Amount to Pay (including tax) is {to_pay} Euros")
    
# Ensure the main function runs if this script is executed
if __name__ == "__main__":
    main()

```
