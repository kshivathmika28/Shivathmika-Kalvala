```python
# TASK1_a
import numpy as np
scores = [61,93,75,67,84,95]
mu = 71
n = 6
dof = n-1
# H0 mu <= 71
# HA mu > 71
# right -tailed t test
alpha = 0.1
sample_mean = np.mean(scores)
sample_mean
```

```python
from scipy.stats import t
import statistics as st
std = st.stdev(scores)
t_stat = (sample_mean-mu)/(std/(n)**0.5)
t_crit = t.ppf((1-0.1),dof)
p_value = t.sf(abs(t_stat),dof)
```

```python
# Task1_b
print(t_crit)
```

```python
# Task1_c
print(t_stat)
print(p_value)
```

```python
# Task1_d
p_value < alpha
```

```python
-t_crit > t_stat
 
```

```python
t_stat > t_crit
```

```python
# reject the null hypothesis means the sciores of the students will be above 71
```

```python
import matplotlib.pyplot as plt
lower_p = sample_mean-4 * std
print(lower_p)
upper_p = sample_mean+4 * std
print(upper_p)
```

```python

graph_values = np.arange(lower_p,upper_p)
plt.plot(graph_values,t.pdf(graph_values, df = dof, loc = sample_mean, scale = std), color='red')
plt.xlabel('t-values')
plt.ylabel('Probability Density')
plt.title('t-Distribution Curve')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# TASK2_a
batch1 = [9,9,9,36,45,48,51,57,69,72,96]
batch2 = [60,39,39,60,79,96,69,60,75,45,90]
std2_1 = st.stdev(batch1)
std2_2 = st.stdev(batch2)
n1 = len(batch1)
n2 = len(batch2)
confidence_level = 0.90
alpha = 1-confidence_level
dof = n1+n2-1
# H0 MU1 = MU2
# HA MU1!= MU2
print(alpha)
```

```python
# Task2_b
alpha_by_2 = alpha/2
t_crit = t.ppf((alpha_by_2),dof)
t_crit
```

```python
s1_mean = st.mean(batch1)
s2_mean = st.mean(batch2)

s1_var = (st.stdev(batch1))**2
s2_var = (st.stdev(batch2))**2
print(s1_var)
print(s2_var)
```

```python
signal = abs(s1_mean - s2_mean)
signal
```

```python
noise = (s1_var * (n1-1) + s2_var * (n2-1))**0.5 * (1/n1 +1/n2)**0.5
noise
```

```python
# Task2_c
t_stats = signal/noise
t_stats
```

```python
# Task2_c
p_value = t.sf(abs(t_stats),dof) * 2
p_value
```
