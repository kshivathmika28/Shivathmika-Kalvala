```python
# Stratified Sampling
grouped_df = df.groupby(["premium"])
depart = df["premium"].unique()
depart
```

```python
samples = []
for mygroup in depart:
    d_data = grouped_df.get_group(mygroup)
    # print(d_data)
    sample_data = d_data.sample(frac = 0.3, random_state = 42)
    # print(sample_data)
    samples.append(sample_data)
# print(samples)
strat_sample_df = pd.concat(samples, ignore_index = True)
strat_sample_df  
```

```python
strat_avg_price = stats.mean(strat_sample_df["price"].tolist())
strat_avg_price
```

```python
# Clustered sampling
nirmal = df.groupby("premium")
mylist = []

for group_name, group_df in nirmal:
    sample_data = group_df.sample(frac=0.2, random_state=4)
    mylist.append(sample_data)

cluster_df = pd.concat(mylist, ignore_index=True)
cluster_df
```

```python
cluster_df_avg = stats.mean(cluster_df["price"].tolist())
cluster_df_avg
```

```python
# TASK3_c

# Take a sample of size 7000
sample_7000 = df.sample(n=5000, random_state=42, replace=True)

# Mean
mean_price = sample_7000["price"].mean()

# Median
median_price = sample_7000["price"].median()

# Mode
try:
    mode_price = sample_7000["price"].mode()[0]  # Can be multiple modes; pick the first
except IndexError:
    mode_price = "No mode"

# Print results
print(f"Mean: {mean_price}")
print(f"Median: {median_price}")
print(f"Mode: {mode_price}")
```
