```python
import pandas as pd
df = pd.read_csv(r'C:\Users\akshi\Documents\C8\DS1_C8_Computers_Data_Project.csv')
df
```

```python
import statistics as stats
avg_price = stats.mean(df["price"])
print(avg_price)
std_price = stats.stdev(df["price"])
print(std_price)
len_d = len(df)
len_d
```

```python
data = df.sample(n = 40, random_state = 42)
data
```

```python
# sample mean of adv_2023  premium computers by randomly selecting 40 computers
import statistics as stats
sample = data[(data["premium"] == "yes")]
budget = sample["ads_2023"]
avg1 = stats.mean(budget)
avg1
```

```python
from scipy.stats import norm
z_crit = norm.ppf(1-0.05)
z_crit
```

```python
z_stats = (avg1-mu)/(sigma/n**0.5)
z_stats
```

```python
p_value = norm.sf(abs(1-alpha))
p_value
```

```python
p_value<alpha
```

```python
z_stats>z_crit
```

```python
# Accept the null hypothesis means advertisng budget spent on promoting premium computers is lesser than or equal to 221.3
```
