```python
import pandas as pd
```

```python
df = pd.read_csv(r'DS1_C5_S5_Hotel Dataset.csv')
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df["hotel"] = df["hotel"].astype("category")
df["arrival_date_month"] = df["arrival_date_month"].astype("category")
df["meal"] = df["meal"].astype("category")
df["country"] = df["country"].astype("category")
df["market_segment"] = df["market_segment"].astype("category")
df["distribution_channel"] = df["distribution_channel"].astype("category")
df["reserved_room_type"] = df["reserved_room_type"].astype("category")
df["assigned_room_type"] = df["assigned_room_type"].astype("category")
df["deposit_type"] = df["deposit_type"].astype("category")
df["customer_type"] = df["customer_type"].astype("category")
df["reservation_status"] = df["reservation_status"].astype("category")
df["reservation_status_date"] = df["reservation_status_date"].astype("category")
```

```python
m = df["agent"].mean()
m1 = df["company"].mean()
df["agent"] = df["agent"].fillna(m)
df["company"] = df["company"].fillna(m1)
```

```python
df.isnull().sum()
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

# TASK1_A

# 1. Distribution of number of special requests
plt.hist(df['total_of_special_requests'], bins=range(0, df['total_of_special_requests'].max()+2), color='skyblue', edgecolor='black')
plt.title('Distribution of Special Requests')
plt.xlabel('Number of Special Requests')
plt.ylabel('Number of Guests')
plt.show()

# TASK1_B
# 2. Scatter plot: special requests vs ADR
plt.scatter(df['total_of_special_requests'], df['adr'], alpha=0.5, color='orange')
plt.title('Special Requests vs ADR')
plt.xlabel('Number of Special Requests')
plt.ylabel('Average Daily Rate (ADR)')
plt.show()

plt.figure(figsize=(6,4))
df['assigned_room_type'].value_counts().plot(kind='hist', color='pink')
plt.title("Room Types Actually Assigned")
plt.xlabel("Room Type Assigned")
plt.ylabel("Number of Guests")
plt.show()


```

```python
# TASK2_A
# Filter for successful bookings (non-canceled)
successful = df[df['is_canceled'] == 0]

# Market segment contribution
plt.figure(figsize=(6,4))
successful['market_segment'].value_counts().plot(kind='bar', color='lightblue')
plt.title('Successful Bookings by Market Segment')
plt.xlabel('Market Segment')
plt.ylabel('Number of Bookings')
plt.show()

# Distribution channel contribution
plt.figure(figsize=(6,4))
successful['distribution_channel'].value_counts().plot(kind='bar', color='lightblue')
plt.title('Successful Bookings by Distribution Channel')
plt.xlabel('Distribution Channel')
plt.ylabel('Number of Bookings')
plt.show()
```

```python
# TASK2_B
plt.figure(figsize=(6,4))
df.groupby('meal')['adr'].mean().plot(kind='bar', color='red')
plt.title('ADR by Meal Type')
plt.xlabel('Meal Type')
plt.ylabel('Average Daily Rate')
plt.show()

# ADR by deposit type
plt.figure(figsize=(6,4))
df.groupby('deposit_type')['adr'].mean().plot(kind='bar', color='orange')
plt.title('ADR by Deposit Type')
plt.xlabel('Deposit Type')
plt.ylabel('Average Daily Rate')
plt.show()

# ADR by required car parking spaces
plt.figure(figsize=(6,4))
df.groupby('required_car_parking_spaces')['adr'].mean().plot(kind='bar', color='blue')
plt.title('ADR by Car Parking Spaces')
plt.xlabel('Car Parking Spaces')
plt.ylabel('Average Daily Rate')
plt.show()
```

```python
# TASK2_C
plt.figure(figsize=(6,4))
plt.hist(df['distribution_channel'], color='cyan', edgecolor='black')
plt.title('Distribution of Bookings by Channel')
plt.xlabel('Distribution Channel')
plt.ylabel('Number of Bookings')
plt.show()
```
