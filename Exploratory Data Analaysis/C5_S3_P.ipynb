```python
# List of columns to analyze
# List of columns to analyze
columns = ["PRICEEACH", "SALES", "MSRP"]

# Initialize dictionaries
mean = {}
median = {}
mode = {}
std = {}
variance = {}
Q1 = {}
Q2 = {}
Q3 = {}
IQR = {}
skew = {}
kurt = {}

# Define the function
def descriptive_analysis_num(df, columns):
    for col in columns:
        mean[col] = df[col].mean()
        median[col] = df[col].median()
        mode[col] = df[col].mode().iloc[0] if not df[col].mode().empty else None
        std[col] = df[col].std()
        variance[col] = df[col].var()
        Q1[col] = df[col].quantile(0.25)
        Q2[col] = df[col].quantile(0.50)
        Q3[col] = df[col].quantile(0.75)
        IQR[col] = Q3[col] - Q1[col]
        skew[col] = df[col].skew()
        kurt[col] = df[col].kurtosis()

        print(f"\n--- {col} ---")
        print(f"Mean: {mean[col]}")
        print(f"Median: {median[col]}")
        print(f"Mode: {mode[col]}")
        print(f"Standard Deviation: {std[col]}")
        print(f"Variance: {variance[col]}")
        print(f"Q1 (25th percentile): {Q1[col]}")
        print(f"Q2 (50th percentile/Median): {Q2[col]}")
        print(f"Q3 (75th percentile): {Q3[col]}")
        print(f"IQR: {IQR[col]}")
        print(f"Skewness: {skew[col]}")
        print(f"Kurtosis: {kurt[col]}")

# Call the function
descriptive_analysis_num(df, columns)

```

```python
# List of columns to analyze
columns = ["PRODUCTLINE", "CUSTOMERNAME", "COUNTRY"]

# Initialize dictionaries
mode = {}
unique = {}
freq_counts = {}
per_distribution = {}

# Define the function
def descriptive_analysis_cat(df, columns):
    for col in columns:
        print(f"\n--- {col} ---")

        # Mode
        mode[col] = df[col].mode().iloc[0] if not df[col].mode().empty else None
        print("Mode:", mode[col])

        # Unique count
        unique[col] = df[col].nunique()
        print("Unique Values:", unique[col])

        # Frequency count
        freq_counts[col] = df[col].value_counts()
        print("Frequency Counts:\n", freq_counts[col])

        # Percentage distribution
        per_distribution[col] = df[col].value_counts(normalize=True) * 100
        print("Percentage Distribution (%):\n", per_distribution[col].round(2))

# Call the function
descriptive_analysis_cat(df, columns)

```

```python
# TASK3
import matplotlib.pyplot as plt
import seaborn as sns
columns = ['PRICEEACH', 'SALES', 'MSRP']
for col in columns:
    print(f"\n--- {col} ---")
    print(f"Mean: {df[col].mean():.2f}")
    print(f"Median: {df[col].median():.2f}")
    print(f"Standard Deviation: {df[col].std():.2f}")
    print(f"Skewness: {df[col].skew():.2f}")
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col], kde=True, bins=30, color='skyblue')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
    bins = pd.cut(df[col], bins=5)
    pie_data = bins.value_counts().sort_index()

    plt.figure(figsize=(6, 6))
    plt.pie(pie_data, labels=pie_data.index.astype(str), autopct='%1.1f%%')
    plt.title(f'Pie Chart of Binned {col}')
    plt.show();
```

```python
# TASK4
columns = ["PRODUCTLINE", "CUSTOMERNAME", "COUNTRY"]

for col in columns:
    print(f"\n--- {col} ---")

    # Mode (most frequent value)
    mode = df[col].mode().iloc[0] if not df[col].mode().empty else None
    print(f"Mode: {mode}")

    # Frequency count
    freq_counts = df[col].value_counts()
    print(f"Frequency Counts:\n{freq_counts}")

    # Percentage distribution
    per_distribution = df[col].value_counts(normalize=True) * 100
    print(f"Percentage Distribution (%):\n{per_distribution.round(2)}")

    # Bar chart (for frequency of categories)
    top_8_categories = freq_counts.head(8)  # Get top 8 categories for the count plot
    plt.figure(figsize=(8, 4))
    sns.countplot(x=df[col], palette='viridis', order=top_8_categories.index)
    plt.title(f'Bar Chart of Top 8 {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.xticks(rotation=45)
    plt.show()

    # Pie chart (binned distribution of categories)
    pie_data = freq_counts.head(8) 
    plt.figure(figsize=(6, 6))
    plt.pie(pie_data, labels=pie_data.index, autopct='%1.1f%%', startangle=90)
    plt.title(f'Pie Chart of {col}')
    plt.show();
```
