```python
df["hotel"] = df["hotel"].astype("category")
df["arrival_date_month"] = df["arrival_date_month"].astype("category")
df["meal"] = df["meal"].astype("category")
df["country"] = df["country"].astype("category")
df["market_segment"] = df["market_segment"].astype("category")
df["distribution_channel"] = df["distribution_channel"].astype("category")
df["reserved_room_type"] = df["reserved_room_type"].astype("category")
df["assigned_room_type"] = df["assigned_room_type"].astype("category")
df["deposit_type"] = df["deposit_type"].astype("category")
df["customer_type"] = df["customer_type"].astype("category")
df["reservation_status"] = df["reservation_status"].astype("category")
df["reservation_status_date"] = df["reservation_status_date"].astype("category")
```

```python
m = df["agent"].mean()
m1 = df["company"].mean()
df["agent"] = df["agent"].fillna(m)
df["company"] = df["company"].fillna(m1)
```

```python
# Task 2.1
import matplotlib.pyplot as plt
import seaborn as sns

monthly_counts = df['arrival_date_month'].value_counts().head(10)

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x=monthly_counts.index, y=monthly_counts.values, palette = sns.color_palette("Set3"))
plt.xlabel('Arrival Date Month')
plt.ylabel('Frequency')
plt.title('Bookings by Month')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show();
```

```python
# Task 2.2
corr = df[['lead_time', 'is_canceled']].corr()

# Plot heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation: lead_time vs is_canceled")
plt.show()
```

# From the above heatmap there is a negative correlation between the lead_time and is_cancelled (blue - shows the negative correlation , red- shows the positive correlation)

```python
# Task3_b & Task3_C
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'], errors='coerce')

# Step 4: Create a 'year_month' column for trend analysis
df['year_month'] = df['reservation_status_date'].dt.to_period('M')

# Step 5: Group by month and calculate cancellation rate
trend = df.groupby('year_month')['is_canceled'].mean()

# Step 6: Plot
trend.plot(kind='line', figsize=(10, 5), title='Cancellation Rate Over Time', marker='o')
plt.ylabel('Cancellation Rate')
plt.xlabel('Year-Month')
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
# Task 4
q1 = df.groupby('customer_type').agg({
    'lead_time': 'mean',
    'total_of_special_requests': 'mean',
    'reserved_room_type': pd.Series.nunique,
    'assigned_room_type': pd.Series.nunique
})

print(q1)

# Question II: Best marketing channels for customer segments
q2_market = pd.crosstab(df['customer_type'], df['market_segment'], normalize='index') * 100

print(q2_market.round(1))


# Question III: Strategy customization based on cancellations, lead time, requests
q3 = df.groupby('customer_type').agg({
    'is_canceled': 'mean',
    'previous_cancellations': 'mean',
    'lead_time': 'mean',
    'total_of_special_requests': 'mean'
})
print("\nQuestion III: Strategy Customization Factors")
print(q3.round(2))
```

```python
# Task6
data = {
    'channel': ['Direct', 'OTA', 'Direct', 'Travel Agent', 'OTA', 'Corporate', 'Direct', 'OTA'],
    'status': ['confirmed', 'canceled', 'confirmed', 'confirmed', 'waiting_list', 'canceled', 'confirmed', 'confirmed'],
    'booking_value': [200, 150, 300, 250, 100, 400, 220, 180],
    'changes': [1, 0, 2, 1, 0, 3, 0, 1],
    'waiting_days': [0, 2, 0, 1, 3, 5, 0, 1],
    'canceled': [0, 1, 0, 0, 0, 1, 0, 0]
}

df = pd.DataFrame(data)

# 1. Effectiveness by channel: conversion rate and average booking value
total = df.groupby('channel').size()
confirmed = df[df['status'] == 'confirmed'].groupby('channel').size()
conversion_rate = (confirmed / total).fillna(0)
avg_value = df.groupby('channel')['booking_value'].mean()

print("Effectiveness by Channel:")
print(pd.DataFrame({'Conversion Rate': conversion_rate, 'Average Booking Value': avg_value}))

# 2. Booking changes vs cancellation rate
changes_vs_cancel = df.groupby('changes')['canceled'].mean()
print("\nCancellation Rate by Number of Changes:")
print(changes_vs_cancel)

# 3. Waiting list duration vs cancellation rate
waiting_vs_cancel = df.groupby('waiting_days')['canceled'].mean()
print("\nCancellation Rate by Waiting Days:")
print(waiting_vs_cancel)

# Optional simple plot for waiting days vs cancellation rate
waiting_vs_cancel.plot(kind='line', marker='o')
plt.title('Cancellation Rate by Waiting List Duration')
plt.xlabel('Waiting Days')
plt.ylabel('Cancellation Rate')
plt.show()
```

```python
# Task7
data = {
    'guest_id': [1,2,3,4,5,1,2,6],
    'is_repeat': [0,0,0,0,0,1,1,0],
    'stay_type': ['business','leisure','business','leisure','business','business','leisure','leisure'],
    'special_request': [0,1,0,1,0,1,0,1],
    'loyalty_member': [0,1,0,0,0,1,1,0],
    'lead_time': [10,5,3,7,20,15,4,2]
}

df = pd.DataFrame(data)

# I. Behavior of repeated guests with stay type
print("Stay type for repeat and non-repeat guests:")
print(df.groupby(['is_repeat', 'stay_type']).size())

# II. Special requests vs loyalty members
print("\nSpecial request rate by loyalty status:")
print(df.groupby('loyalty_member')['special_request'].mean())

# III. Lead time distribution for repeat vs non-repeat
print("\nAverage lead time:")
print(df.groupby('is_repeat')['lead_time'].mean())

# Optional: show lead time comparison as boxplot
df.boxplot(column='lead_time', by='is_repeat')
plt.title('Lead Time: Repeat vs Non-Repeat')
plt.suptitle('')
plt.xlabel('Repeat Guest (0 = No, 1 = Yes)')
plt.ylabel('Lead Time (days)')
plt.show()
```
