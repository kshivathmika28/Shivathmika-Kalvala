```python
import pandas as pd
```

```python
df = pd.read_csv(r'DS1_C5_S1_EmployeeAttrition_Data_Practice.csv')
df
```

```python

df.isnull().sum()
```

```python
# STEP1
df["BusinessTravel"] = df["BusinessTravel"].fillna("Not Available")
df["EducationField"] = df["EducationField"].fillna("Not Available")
df["Gender"] = df["Gender"].fillna("Not Available")
df["JobRole"] = df["JobRole"].fillna("Not Available")
df["MaritalStatus"] = df["MaritalStatus"].fillna("Not Available")
df["Over18"] = df["Over18"].fillna("Not Available")
df["Age"] = df["Age"].fillna("0")
df["Department"] = df["Department"].fillna("Not Available")


```

```python
df.isnull().sum()
```

```python
# STEP2
df["Age"] = df["Age"].astype(float)
df["Attrition"] = df["Attrition"].astype('category')
df["BusinessTravel"] = df["BusinessTravel"].astype('category')
df["Department"] = df["Department"].astype('category')
df["Gender"] = df["Gender"].astype('category')
df["JobRole"] = df["JobRole"].astype('category')
df["MaritalStatus"] = df["MaritalStatus"].astype('category')
df["Over18"] = df["Over18"].astype('category')
df["OverTime"] = df["OverTime"].astype('category')
df["EducationField"] = df["EducationField"].astype('category')
```

```python
# TASK1.1
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(x='BusinessTravel', hue='Attrition', data=df)
plt.title('Employee Attrition by Business Travel Category')
plt.xlabel('Business Travel Frequency')
plt.ylabel('Number of Employees')
plt.legend(title='Attrition')
plt.show()


```

# In the business travel category Travel_Rarely has the highest attrition rate

```python
# TASK1.2
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
grouped_data = df.groupby(['EducationField', 'Attrition']).size().reset_index(name='Count')

plt.figure(figsize=(10, 6))
sns.barplot(x='EducationField', y='Count', hue='Attrition', data=grouped_data)
plt.title('Employee Attrition by Education Field')
plt.xlabel('Education Field')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45)
plt.legend(title='Attrition')
plt.tight_layout()
plt.show();
```

# Life sciences field  has the highest attrition rate

```python
# TASK3
filtered_df = df[(df['Attrition'] == 'Yes') & (df['BusinessTravel'] == 'Travel_Rarely')]
attrition_counts = filtered_df['JobRole'].value_counts()
plt.figure(figsize=(10, 6))
plt.barh(attrition_counts.index, attrition_counts.values, color='orange')
plt.xlabel('Count of Attrited Employees (Traveling Rarely)', fontsize=12)
plt.ylabel('JobRole', fontsize=12)
plt.title('Attrited Employees by JobRole (Traveling Rarely)', fontsize=14)
plt.show();
```

# Labaratory Technician dept has the highest count of attrited employees

```python
# TASK4
filtered_df = df[
    (df['Attrition'] == 'Yes') &
    (df['EducationField'].isin(['Life Sciences', 'Medical']))
]
attrition_counts = filtered_df['JobRole'].value_counts()
plt.figure(figsize=(10, 6))
plt.barh(attrition_counts.index, attrition_counts.values, color='violet')
plt.xlabel('Count of Attrited Employees ', fontsize=12)
plt.ylabel('JobRole', fontsize=12)
plt.title('Attrited Employees by JobRole', fontsize=14)
plt.show();
```
