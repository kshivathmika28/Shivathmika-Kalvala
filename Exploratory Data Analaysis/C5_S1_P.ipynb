```python
# STEP1
# Fill missing values for categorical columns using the first mode
# Fill missing numerical value
df['Age'].fillna(df['Age'].mean())

# Fill missing categorical values with mode
categorical_columns = ['BusinessTravel', 'Department', 'EducationField', 'Gender',
                       'JobRole', 'MaritalStatus', 'Over18']

for col in categorical_columns:
    df[col].fillna(df[col].mode()[0])

# To verify all missing values are handled
print(df.isnull().sum())

```

```python
# STEP2
df["Age"] = df["Age"].astype(float)
df["Attrition"] = df["Attrition"].astype('category')
df["BusinessTravel"] = df["BusinessTravel"].astype('category')
df["Department"] = df["Department"].astype('category')
df["Gender"] = df["Gender"].astype('category')
df["JobRole"] = df["JobRole"].astype('category')
df["MaritalStatus"] = df["MaritalStatus"].astype('category')
df["Over18"] = df["Over18"].astype('category')
df["OverTime"] = df["OverTime"].astype('category')
df["EducationField"] = df["EducationField"].astype('category')
```

```python
# TASK2
from pandas.api.types import is_numeric_dtype, is_categorical_dtype
list_num = []
list_cat = []
def check_dtype(df):
    for col in df.columns:
        if is_numeric_dtype(df[col]):
            list_num.append(col)
        else:
            list_cat.append(col)
    return list_num, list_cat
check_dtype(df)
print("Numerical columns:",list_num)
print("Categorical columns:", list_cat)
```

```python
# TASK3
def cat_col(df, list_cat):
    for col in list_cat:
        unique = df[col].nunique()
        mode = df[col].mode()[0]
        missing = df[col].isnull().sum()

        print(f"\nColumn: {col}")
        print(f"Unique values: {unique}")
        print(f"Mode: {mode}")
        print(f"Missing values: {missing}")

        if missing > 0:
            df[col].fillna(mode, inplace=True)
            print(f"Missing values filled with mode: {mode}")
        else:
            print("No missing values to fill.")
    return df
```

```python
# TASK5
def descrptive_analysis(df,list_num):
    for i in list_num:
        mean1 = df[i].mean()
        median1 = df[i].median()
        std1 = df[i].std()
        missing = df[i].isnull().sum()
        s = df[i].skew()
        print(i)
        print(f"Mean value is: {mean1}")
        print(f"Median value is: {median1}")
        print(f"STD value is: {std1}")
        print(f"Total number of missing values: {missing}")
        print(f"skewness: {s}")
        print("---------------------------------------------")
descrptive_analysis(df,list_num)
        
```
