```python
import pandas as pd
```

```python
list = pd.read_html(r"https://www.basketball-reference.com/leagues/NBA_2021_per_game.html")
df = list[0]
df
```

```python
# Task 2 Extract the following columns from the dataset to perform the analysis:a) G b) MP c) FG d) PTS 
# e) TRB f) AST g) STL h) BLK i) TOV

import pandas as pd

# Extract the specified columns
selected_columns = df[["G", "MP", "FG", "PTS", "TRB", "AST", "STL", "BLK", "TOV"]]

# Print the selected columns
print(selected_columns)
```

```python
# Task3: Extract the following columns from the dataset to perform the analysis:a) Gb) MPc) FGd) PTS
# Import required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import kurtosis

# Use the already extracted columns
selected_columns = df[['G', 'MP', 'FG', 'PTS']]

# Calculate mean for each column
print("Mean Values:")
print(selected_columns.mean())

# Calculate median for each column
print("\nMedian Values:")
print(selected_columns.median())

# Calculate kurtosis for each column
print("\nKurtosis Values:")
print(selected_columns.apply(lambda x: kurtosis(x.dropna())))

# Create a summary DataFrame with all statistics
summary_stats = pd.DataFrame({
    'Mean': selected_columns.mean(),
    'Median': selected_columns.median(),
    'Kurtosis': selected_columns.apply(lambda x: kurtosis(x.dropna()))
})

print("\nSummary Statistics:")
print(summary_stats.round(2))

# Create box plots for all columns
plt.figure(figsize=(12, 6))
sns.boxplot(data=selected_columns)
plt.title('Distribution of G, MP, FG, and PTS')
plt.ylabel('Values')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Create individual box plots for better visualization
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Individual Box Plots for Each Column')

# Plot each column in a separate subplot
sns.boxplot(y=selected_columns['G'], ax=axes[0,0])
axes[0,0].set_title('Games (G)')
axes[0,0].grid(True, linestyle='--', alpha=0.7)

sns.boxplot(y=selected_columns['MP'], ax=axes[0,1])
axes[0,1].set_title('Minutes Played (MP)')
axes[0,1].grid(True, linestyle='--', alpha=0.7)

sns.boxplot(y=selected_columns['FG'], ax=axes[1,0])
axes[1,0].set_title('Field Goals (FG)')
axes[1,0].grid(True, linestyle='--', alpha=0.7)

sns.boxplot(y=selected_columns['PTS'], ax=axes[1,1])
axes[1,1].set_title('Points (PTS)')
axes[1,1].grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.subplots_adjust(top=0.9)
plt.show()

# Optional: Create violin plots for more detailed distribution visualization
plt.figure(figsize=(14, 7))
sns.violinplot(data=selected_columns)
plt.title('Distribution of G, MP, FG, and PTS (Violin Plots)')
plt.ylabel('Values')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()





```
