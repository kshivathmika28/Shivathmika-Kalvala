```python
#Task1
data.isnull().sum()
```

```python
a = data['gender'].value_counts()
b = data['test1'].value_counts()
c = data['test2'].value_counts()
print(a)
print(b)
print(c)
```

```python
# Replace missing gender values with the most frequent one
most_common_gender = data['gender'].mode()[0]
data['gender'] = data['gender'].fillna(most_common_gender)

# Replace missing test1 and test2 values with their respective column means
data['test1'] = data['test1'].fillna(data['test1'].mean())
data['test2'] = data['test2'].fillna(data['test2'].mean())
```

```python
#task2
data.duplicated()
```

```python
data_clean = data.drop_duplicates()
data_clean
```

```python
#task3
data["tuitionfee_waiver"] = data["tuitionfee_waiver"].str.replace("Does not qualify","No",regex = False)
data["tuitionfee_waiver"] = data["tuitionfee_waiver"].str.replace("Qualifies for reduced/free","Yes",regex = False)
data
```

```python
#task4
grp_by = data.groupby(by= ["school"])["test1"].mean()
print(grp_by)
max_score = max(grp_by)
print(f"The maximum average score in test1 is {max_score}")
```

```python
#task5

max_Score = data.groupby(by=["school"])["test2"].max()
print(f"Themaximum score in test2 is{max_Score}")
pvt = data.pivot_table(index = ["school"],values = "test2", aggfunc = "max")
print(pvt.head(5))
```

```python
#Task6
eligible_students = data[
    (data['classroom'] == '5LQ') &
    (data['test2'] >= 80) &
    (data['school'] == 'FBUMG')
].copy()

scholarship = pd.Series([500] * len(eligible_students), index=eligible_students.index)

eligible_students = pd.concat([eligible_students, scholarship.rename('scholarship_amount')], axis=1)

columns_to_display = ['student_id', 'school', 'classroom', 'gender', 'test2', 'scholarship_amount']
print(eligible_students[columns_to_display])

```
