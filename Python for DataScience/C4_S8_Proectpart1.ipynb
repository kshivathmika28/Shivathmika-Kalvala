```python
# Task4 Calculate kurtosis for the extracted columns: Runs, BF, 4s, and 6s
def calculate_kurtosis(df):
    """
    Calculate kurtosis for specified cricket statistics columns
    """
    columns = ['Runs', 'BF', '4s', '6s']
    kurtosis_values = {}
    
    for col in columns:
        kurtosis_values[col] = df[col].kurtosis()
        
    return kurtosis_values

 

```
